### Desarrolle una funcion la cual reciba dos parametros, una imagen y un entero llamado color, la funcion debe devolver una imagen la cual tenga activos los canales de color segun los siguientes


```python
def procesamiento_imagen(image, color):
    width = image.shape[1]
    height = image.shape[0]
    new_image = np.zeros((height, width, 3))
    option_valid = True
    for i in range(0, height):
        for j in range(0, width):
            pixel = image[i,j]
            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]
            if color == 1:                
                new_image[i,j] = [blue, 0, 0]
            elif color == 2:
                new_image[i,j] = [0, green, 0]
            elif color == 3:
                new_image[i,j] = [0, 0, red]
            elif color == 10:
                new_image[i,j] = [0, green, red]
            elif color == 20:
                new_image[i,j] = [blue, green, 0]   
            elif color == 30:
                new_image[i,j] = [blue, 0, red]
            else:
                new_image = image
                option_valid = False
    if option_valid:
        cv2.imwrite("new_image.jpg", new_image)
        new_image = cv2.imread("new_image.jpg")
        new_image = cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB)
    else: 
        print("Color no implementado ")
    return new_image
    
```

### Funcion de reconstruccion de imagen gris

```python
def reconstruir_imagen(img_gray_blue, img_gray_green, img_gray_red  ):
    width = img_gray_blue.shape[1]
    height = img_gray_blue.shape[0]
    new_image = np.zeros((height, width, 3))
    option_valid = True
    for i in range(0, height):
        for j in range(0, width):
            blue = img_gray_blue[i,j][0]
            green = img_gray_green[i,j][0]
            red = img_gray_red[i,j][0]
            new_image[i,j] = [blue, green, red]
    cv2.imwrite("new_image_final.jpg", new_image)
    new_image = cv2.imread("new_image_final.jpg")
    new_image = cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB)
    return new_image
```

### Problema 3:

#### Cree una funci´on que dada una imagen cree una escala de grises en tres dimensiones, tome en cuenta que su funci´on debe crear 3 im´agenes como salida. Para entregar este ejercicio debe incluir una las im´agenes que haya utilizado como prueba y el resultado de las misma, no puede utilizar la imagen del Problema #2

```python
def write_and_show(strImg, arrImg):
    cv2.imwrite(strImg, arrImg)
    temp = cv2.imread(strImg)
    temp = cv2.cvtColor(temp, cv2.COLOR_BGR2RGB)
    plt.imshow(temp)
    
def convertir_escala_grises(image):
    width = image.shape[1]
    height = image.shape[0]
    option_valid = True
    gray_green = np.zeros((height, width, 1)) 
    gray_blue = np.zeros((height, width, 1)) 
    gray_red = np.zeros((height, width, 1)) 
    for i in range(0, height):
        for j in range(0, width):
            pixel = image[i,j]
            gray_blue[i,j] = pixel[2]
            gray_green[i,j] = pixel[1]
            gray_red[i,j] = pixel[0]
    write_and_show("gray_red.jpg", gray_red)
    write_and_show("gray_green.jpg", gray_green)
    write_and_show("gray_blue.jpg", gray_blue)
    
    
```

```python
image_test =  cv2.imread("imagen_test.jpg")
image_test_correct = cv2.cvtColor(image_test, cv2.COLOR_BGR2RGB)
convertir_escala_grises(image_test_correct)
```

### Problema 6:

HSV es un modelo de color cilíndrico que reasigna los colores primarios RGB en dimensiones que son más fáciles de entender para los humanos. Al igual que el sistema de color Munsell, estas dimensiones son tono, saturación y valor.

Hue especifica el ángulo del color en el círculo de color RGB. Un tono de 0 ° da como resultado rojo, 120 ° da como resultado verde y 240 ° da como resultado azul.
La saturación controla la cantidad de color utilizada. Un color con un 100% de saturación será el color más puro posible, mientras que un 0% de saturación producirá una escala de grises.
Value controla el brillo del color. Un color con 0% de brillo es negro puro, mientras que un color con 100% de brillo no tiene negro mezclado con el color. Debido a que esta dimensión a menudo se denomina brillo, el modelo de color HSV a veces se denomina HSB, incluso en P5.js.

```python
img = cv2.imread('imagen_test.jpg')

# image to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
# image to HSV
img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

plt.imshow(img)
```
